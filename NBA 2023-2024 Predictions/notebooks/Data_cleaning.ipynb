```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
adv_df = pd.read_csv('../data/raw/Advanced.csv')
player_df = pd.read_csv('../data/raw/Player Totals.csv')
```

```python
different_cols = player_df.columns.difference(adv_df.columns)
```

```python
player2 = player_df[different_cols]
```

```python
player_data = pd.merge(adv_df, player2, left_index=True, right_index=True, how='inner')
```

```python
player_data.head()
```

```python
def describe_dataframe(df):
    print("  TYPES   ".center(82, '-'))
    print(df.dtypes)

    print("  SHAPE   ".center(82,'-'))
    print(f"Rows: {df.shape[0]}")
    print(f"Columns: {df.shape[1]}")

    print("  DUPLICATES  ".center(82,'-'))
    print(df.duplicated().sum())

    print("  NaN VALUES".center(82,'-'))
    print(df.isna().sum())

    print("  QUANTILES   ".center(82,'-'))
    print(df.quantile([0,0.05,0.50,0.95,0.99,1]).T)

    print("  DATAFRAME HEAD      ".center(82, '-'))
    print(df.head(n=5))
    print("  DATAFRAME TAIL      ".center(82, '-'))
    print(df.tail(n=5))
```

```python
describe_dataframe(player_data)
```
